Create Range

```python
def income_range(salary):
    if(salary<=25000):
        return'0-25000'
    elif(salary<=50000):
        return'50000-750000'
    elif(salary<=100000):
        return'750000-100000'
    elif(salary<=125000):
        return'100000-125000'
    elif(salary<=150000):
        return'125000-150000'
    elif(salary<=175000):
        return'150000-175000'
    elif(salary<=200000):
        return'175000-200000'
    elif(salary<=225000):
        return'200000-225000'
    elif(salary<=250000):
        return'225000-225000'
    elif(salary<=275000):
        return'250000-275000'
    elif(salary<=300000):
        return'275000-300000'
    else:
        return'300000-above'
```

```python
df['income_range']=df['amt_income_total'].apply(income_range)
df
```

```python
o=df['income_range'].value_counts()
idf=pd.DataFrame(o)
idf=idf.reset_index()
idf
```

```python
createbar(x=o.index,y=o.values,xlabe='income_range',ylabe='count',labe='income_range',size=(20,5))
```

```python
sns.barplot(data=idf,y='income_range',x='count')
plt.figure(figsize=(20,5))
plt.show()
```
